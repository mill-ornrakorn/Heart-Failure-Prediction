```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# โหลดข้อมูลที่ clean แล้ว
df = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/Heart-Failure-Prediction/heart_failure_clinical_records.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# ตรวจสอบ na ในข้อมูล
df.isna().sum()
```

```python
# เปลี่ยน data type ของ column
df['anaemia'] = df['anaemia'].astype(str)
df['diabetes'] = df['diabetes'].astype(str)
df['high_blood_pressure'] = df['high_blood_pressure'].astype(str)
df['sex'] = df['sex'].astype(str)
df['smoking'] = df['smoking'].astype(str)
df['DEATH_EVENT'] = df['DEATH_EVENT'].astype(str)
```

```python
df.info()
```

```python
df_cat  = pd.get_dummies(df.select_dtypes('object') , drop_first=True)
df_cat
```

```python
df_cat.info()
```

```python
df_num = df.select_dtypes('number')
df_num
```

```python
df_num.info()
```

```python
# from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df_num_scaler = MinMaxScaler().fit_transform(df_num)
df_num_scaler
```

```python
df_num_scaler_final = pd.DataFrame(df_num_scaler, columns=df_num.columns, index=df_num.index)
df_num_scaler_final
```

```python
df_final = pd.concat([df_num_scaler_final, df_cat],axis=1) # axis=1 จะเป็นการต่อในแนวนอน
df_final
```

```python
df_final.to_csv('heart_failure_clinical_records-scaled.csv', index=False)
```
